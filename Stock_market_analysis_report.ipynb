```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_excel("D:\python\Excel\yahoo_data.xlsx")
```

```python
df=pd.DataFrame(data)
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Basic cleaning operations
df.dropna()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.drop_duplicates()
```

```python
#Average opening market value by date
result=df.groupby('Date')[['Open']].sum().mean()
print("Average Opening stock value",result)
```

```python
#Average close market value by date
result=df.groupby('Date')[['Close*']].sum().mean()
print("Average Closing stock value",result)
```

```python
max_index = df["Volume"].idxmax() # Returns 1 (index of the max value)
highest_branch = df.loc[max_index] # Fetches the row with max amount

print("Highest volume:\n",max_index)
print("Highest value branch:\n",highest_branch)
```

```python
max_index = df['Open'].idxmax() # Returns 1 (index of the max value)
highest_Opening = df.loc[max_index]
print("Highest price details:\n",highest_Opening)
```

```python
df.head(5)
```

```python
# Basic visualizations
plt.hist(df['Open'])      
plt.show()# Histogram
```

```python

# If you want separate histograms:
df[['Open', 'Close*']].hist(bins=20, layout=(1, 2), figsize=(12, 4))
plt.suptitle('Histograms of Open and Close* Prices')
plt.show()


```

```python
# Or overlay both series in a single plot with transparency:
plt.figure(figsize=(8, 5))
plt.hist(df['Open'].dropna(), bins=30, alpha=0.5, label='Open')
plt.hist(df['Close*'].dropna(), bins=30, alpha=0.5, label='Close*')
plt.legend()
plt.title('Histogram of Open vs. Close*')
plt.show()
```

```python

plt.figure(figsize=(12, 5))
plt.bar(df['Open'], df['Volume'])
plt.title("Open and Volume")
plt.xlabel("Open")
plt.ylabel("Volume")
plt.show()
```

```python

plt.figure(figsize=(12, 5))
plt.bar(df['Open'], df['Close*'])
plt.title("Open and Close Stock Value")
plt.xlabel("Open")
plt.ylabel("Close")
plt.show()
```

```python
x = df.corr(numeric_only=True)
print(x)
```

```python
sns.heatmap(x, annot=True, cmap='viridis')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
